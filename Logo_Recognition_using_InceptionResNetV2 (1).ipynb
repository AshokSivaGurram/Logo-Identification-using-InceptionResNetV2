```python
#Importing all required libraries
import tensorflow as tf
import cv2
import random 
import numpy as np
import matplotlib.pyplot as plt
import os
import tensorflow as tf
from tensorflow.keras.applications import *
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
from tensorflow.keras.utils import load_img
```

```python
lt = [cv2.ROTATE_180,cv2.ROTATE_90_COUNTERCLOCKWISE,cv2.ROTATE_90_CLOCKWISE]
def brightness(img):
    value = random.uniform(0.5, 2)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv = np.array(hsv, dtype = np.float64)
    hsv[:,:,1] = hsv[:,:,1]*value
    hsv[:,:,1][hsv[:,:,1]>255]  = 255
    hsv[:,:,2] = hsv[:,:,2]*value 
    hsv[:,:,2][hsv[:,:,2]>255]  = 255
    hsv = np.array(hsv, dtype = np.uint8)
    img = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    return img
```

```python
#assigning directory
import pathlib
import glob
directory=pathlib.Path("/content/drive/MyDrive/logos1")
```

```python
resultant="/content/augmentedimages2"
```

```python
items = os.listdir(directory)

classes=[]
count=0
images = []
labels=[]
for i in items:
    i1 = 0
    print(i)
    classes.append(i)
    path1 = f"{directory}/{i}"
    a = random.randint(5,10)
    img = cv2.imread(path1)
    img = cv2.resize(img,(299,299))
    k = i.split(".")[0]
    cv2.imwrite(f"{resultant}\{k}{i1}.jpg",img)
    
    i1+=1
    while a!=0:
        img = cv2.rotate(img,lt[random.randint(0,2)])
        images.append(img)
        cv2.imwrite(f"{resultant}\{k}{i1}.jpg",img)
        i1+=1
        labels.append(count)
        if a%2==0:
            img = brightness(img)
            images.append(img)
            cv2.imwrite(f"{resultant}\{k}{i1}.jpg",img)
            i1+=1
            labels.append(count)
        a-=1
    count+=1
images = np.array(images) 
labels = np.array(labels)
```

```python
images.shape
```

```python
# change the model here for alternatives

model = InceptionResNetV2(weights="imagenet")
for i in model.layers:
    i.trainable =  False
```

```python
len(model.layers)
```

```python
model.summary()
```

```python
model2 = Sequential() 

# adding layers from pre-trained model
for i in range(18):
    model2.add(model.layers[i])
```

```python
# adding custum layers
model2.add(Flatten())
model2.add(Dense(400,activation="relu"))
model2.add(Dense(350,activation="relu"))
model2.add(Dense(309,activation="Softmax"))
```

```python
model2.summary()
```

```python
import tensorflow as tf
class myCallback(tf.keras.callbacks.Callback): 
    def on_epoch_end(self, epoch, logs={}): 
        print("call")
        if(logs.get('accuracy') > .85):   
            print("\nReached %2.2f%% accuracy, so stopping training!!" %(85))   
            self.model.stop_training = True
callbacks = myCallback()
```

**Training and Validation**

```python
# model1.summary()
model2.compile(optimizer="Adadelta",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
model2.fit(images,labels,epochs=100,callbacks=[callbacks])
```

```python
model2.evaluate(images,labels)
```

```python
def predict(i,model,labels):
    path1 = "/content/drive/MyDrive/Testing Set/android.jpg"
    img = cv2.imread(path1)
    img = cv2.resize(img,(299,299))
    a = np.argmax(model.predict(np.array([img])))
    img = cv2.putText(img, labels[a], (-5,20), cv2.FONT_HERSHEY_SIMPLEX,1, (200,200,0), 3, cv2.LINE_AA)
    plt.imshow(img)
```
